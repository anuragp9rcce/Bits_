```python
import pyarrow as pa
import pandas as pd
from datasets import Dataset

# Sample pandas DataFrame
df = pd.read_csv(r"C:\Users\ANURAG\Desktop\imdb_bits\intent_classification.csv")
from sklearn.model_selection import train_test_split

# Assuming your dataset is stored in a pandas DataFrame called 'df'
# and the target variable is in the 'lable' column.
train_df, test_df = train_test_split(
    df, test_size=0.2, stratify=df["label"], random_state=42
)
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# Convert pandas DataFrame to Arrow Table
train_ds = Dataset.from_pandas(train_df)
test_ds = Dataset.from_pandas(test_df)
print(train_ds)
print(test_ds)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compute_metrics(y_pred,y_test):
    print(type(y_test))
    labels = y_test
    print(type(labels))
    print(labels)
    print(type(y_test))
    preds = y_pred.cpu().tolist()
    print(type(preds))
    print(preds)


    # Calculate accuracy
    accuracy = accuracy_score(labels, preds)

   # Calculate precision, recall, and F1-score
    precision = precision_score(labels, preds, average='weighted')
    recall = recall_score(labels, preds, average='weighted')
    f1 = f1_score(labels, preds, average='weighted')

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
# Load SetFit model from Hub
model = SetFitModel.from_pretrained("sentence-transformers/paraphrase-mpnet-base-v2")

# Create trainer
trainer = SetFitTrainer(
    model=model,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    loss_class=CosineSimilarityLoss,
    batch_size=16,
    metric = compute_metrics,
    num_iterations=20, # Number of text pairs to generate for contrastive learning
    num_epochs=1 # Number of epochs to use for contrastive learning
)


```

```python
# only essay feature is given in data, project_title  feature is not  available in
# given data
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf= TfidfVectorizer(min_df=10, ngram_range=(1,2),max_features=10000)
tfidf.fit(train_df["text"].values)
X_train_text_tdidf = tfidf.transform(train_df["text"].values)
X_test_text_tfidf  =tfidf.transform(test_df["text"].values)
print(X_train_text_tdidf.shape)
print(X_test_text_tfidf.shape)
```

```python
# stroing word:idf mapping in dictionary
dictionary = dict(zip(tfidf.get_feature_names(), list(tfidf.idf_)))
# storing vocab in a set
tfidf_words = set(tfidf.get_feature_names())
```

```python
# vectorization of text data.# only essay feature is given in data, project_title  feature is not  available in  given data
import numpy as np

import pickle
with open(r"D:\ML\ML AND DL\APPLIED AI ASSIGNMENTS\11_Donors_choose_GBDT\glove_vectors", 'rb') as f:
    model = pickle.load(f)
    glove_words =  set(model.keys())


def get_tfidf_w2v_vectors(X,dictionary,glove_words,tfidf_words,model) :
    from tqdm import tqdm
    tfidf_w2v_vectors = []
    for sentence in tqdm(X["text"].values): # for each review/sentence
        vector = np.zeros(300) # as word vectors are of zero length
        tf_idf_weight =0; # num of words with a valid vector in the sentence/review
        for word in sentence.split(): # for each word in a review/sentence
            if (word in glove_words) and (word in tfidf_words):
                vec = model[word] # getting the vector for each word
                # here we are multiplying idf value(dictionary[word]) and the tf value((sentence.count(word)/len(sentence.split())))
                tf_idf = dictionary[word]*(sentence.count(word)/len(sentence.split())) # getting the tfidf value for each word
                vector += (vec * tf_idf) # calculating tfidf weighted w2v
                tf_idf_weight += tf_idf
        if tf_idf_weight != 0:
            vector /= tf_idf_weight
        tfidf_w2v_vectors.append(vector)
    return np.array(tfidf_w2v_vectors)

X_train_essay_tfidf_w2v_vectors = get_tfidf_w2v_vectors(train_df,dictionary,glove_words,tfidf_words,model)
X_test_essay_tfidf_w2v_vectors = get_tfidf_w2v_vectors(test_df,dictionary,glove_words,tfidf_words,model)

print(X_train_essay_tfidf_w2v_vectors.shape)
print(X_test_essay_tfidf_w2v_vectors.shape)
```

```python
from hyperopt import fmin, tpe, hp, Trials, STATUS_OK
import xgboost as xgb
from sklearn.model_selection import cross_val_score
import numpy as np

# Define the objective function
def objective(params):
    # Convert float parameters to integers for XGBoost
    params['max_depth'] = int(params['max_depth'])
    params['n_estimators'] = int(params['n_estimators'])

    # Create the model
    model = xgb.XGBClassifier(
        n_jobs=-1,
        use_label_encoder=False,
        eval_metric="logloss",
        **params
    )

    # Perform cross-validation
    cv_score = cross_val_score(
        model,
        X_train_essay_tfidf_w2v_vectors,
        Y_train,
        cv=5,
        scoring="neg_log_loss"
    )

    # Objective is to minimize negative log loss
    mean_cv_score = -np.mean(cv_score)
    return {'loss': mean_cv_score, 'status': STATUS_OK}

# Define the hyperparameter search space
search_space = {
    'max_depth': hp.choice('max_depth', [2, 4, 8]),
    'n_estimators': hp.quniform('n_estimators', 100, 300, 50),
    'learning_rate': hp.uniform('learning_rate', 0.01, 0.3),
    'subsample': hp.uniform('subsample', 0.5, 1.0),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1.0),
}

# Initialize trials object to store results
trials = Trials()

# Run hyperparameter optimization
best_params = fmin(
    fn=objective,                # Objective function
    space=search_space,          # Search space
    algo=tpe.suggest,            # Tree of Parzen Estimators algorithm
    max_evals=50,                # Number of evaluations
    trials=trials,               # Record of trials
    verbose=True
)

# Print the best parameters
print("Best Parameters:", best_params)

```
